Seperate Train, Val and Test set according to Year

```python
year=pd.to_datetime(raw_df.Date).dt.year
```

```python
train=raw_df[year<2015]
val=raw_df[year==2015]
test=raw_df[year>2015]
```

```python
test.head()
```

Seperate X, y columns

```python
inputcol=list(train.columns)[1:-1]
outcol=["RainTomorrow"]
```

```python
train_x=train[inputcol].copy()
train_y=train[outcol].copy()
```

```python
test_x=test[inputcol].copy()
test_y=test[outcol].copy()
```

```python
val_x=val[inputcol].copy()
val_y=val[outcol].copy()
```

Seperate num and cat columns

```python
numcol=train_x.select_dtypes(include=np.number).columns.tolist()[:-1]
catcol=train_x.select_dtypes("object").columns.tolist()
```

```python
catcol
```

```python
train_x[numcol].info()
```

Fill null values with mean

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy = 'mean')
train_x[numcol].isna().sum()
```

```python
imputer.fit(raw_df[numcol])

train_x[numcol] = imputer.transform(train_x[numcol])
val_x[numcol] = imputer.transform(val_x[numcol])
test_x[numcol] = imputer.transform(test_x[numcol])
```

```python
train_x.isnull().sum()
```

Scale num values

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(raw_df[numcol])
```

```python
train_x[numcol] = scaler.transform(train_x[numcol])
val_x[numcol] = scaler.transform(val_x[numcol])
test_x[numcol] = scaler.transform(test_x[numcol])
```

```python
train_x.describe()
```

```python
train_x[numcol].nunique()
```

```python
raw_df[catcol].nunique()
```

```python
from sklearn.preprocessing import OneHotEncoder
enc=OneHotEncoder(sparse=False, handle_unknown='ignore')
enc.fit(raw_df[catcol])
```

```python
enc.categories_
```

```python
encoded_cols = list(enc.get_feature_names(catcol))
print(encoded_cols)
```

```python
train_x[encoded_cols] = enc.transform(train_x[catcol])
val_x[encoded_cols] = enc.transform(val_x[catcol])
test_x[encoded_cols] = enc.transform(test_x[catcol])
```

```python
print('train_inputs:', train_x.shape)

print('val_inputs:', val_x.shape)

print('test_inputs:', test_x.shape)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='liblinear')
model.fit(train_x[numcol + encoded_cols], train_y.values.ravel())
```

```python
X_train = train_x[numcol + encoded_cols]
X_val = val_x[numcol + encoded_cols]
X_test = test_x[numcol + encoded_cols]
```

```python
prediction = model.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(train_y, prediction)
```

```python
prediction = model.predict(X_val)
accuracy_score(val_y, prediction)
```

```python
prediction = model.predict(X_test)
accuracy_score(test_y, prediction)
```
